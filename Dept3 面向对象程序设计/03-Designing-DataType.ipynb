## 3.3 设计数据类型的原则

我们再回顾之前学习的三个概念：
- 客户端
- API
- 实现

我们在着手于数据类型设计时，应该以什么样的顺序，或者是采用什么样的思路去设计呢？ 这之中有没有一些前人留下的可参考的东西呢？

答案是：Yes。

在上学期 OOP实验课中，第一节课就讲到了，我们要去设计一个数据模型，还有消息驱动的概念，要以数据为中心，先设计好一个数据模型，再想想如何去“触发”... 

这些虽然是在讲解 MFC 的时候提到的，但是现在想想，在 OOP中，准确说我们需要设计一个自己的数据类型时，也应该采用这样的思路。

`->>` 

**编写程序的首要任务就是尽量理解我们需要的数据类型，理解的过程就是一个设计任务。 通常，构建软件最重要的和最具挑战的就是设计API。**

## 3.4 API 设计原则 

1. 开发一个数据类型的主要目的就是简化客户端代码，所以说我们可以先从客户端角度出发，看看我们需要什么功能， **先写客户端代码，再写实现代码** 。

2. API 设计“太窄”、“太宽”、“太通用”、“太专用”都不行，牢牢把握一个原则：**为客户端提供所需方法，仅此而已。**
