## 3.4 API 设计原则 

1. 开发一个数据类型的主要目的就是简化客户端代码，所以说我们可以先从客户端角度出发，看看我们需要什么功能， **先写客户端代码，再写实现代码** 。

2. API 设计“太窄”、“太宽”、“太通用”、“太专用”都不行，牢牢把握一个原则：**为客户端提供所需方法，仅此而已。**

## 3.7 继承

继承，在今年年初 OOP 实验课上，老师有提到 “基类”、“类”，这些术语和 继承 有关系。

继承这一节，如何去学习、如何去理解？下面通过一个动物分类的例子去理解。

**父类 (Parent Class)**：父类是指被其他类继承的类，它可以有自己的属性和方法，而且也可以被其他类继承。

**子类 (Subclass)**：子类是从父类继承而来的类，它继承了父类的属性和方法，并且可以添加自己的属性和方法。


```python
# 父类
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        pass
```

```python
# 子类
# class Animal:
#     def __init__(self, name):
#         self.name = name

#     def speak(self):
#         pass

class Dog(Animal):
    def speak(self):
        return "Woof!"
```

在这个例子中，Dog 类是一个子类，它继承了 Animal 类的属性和方法，并且添加了自己的 speak 方法来表示狗的叫声。 

**父类被子类继承，子类可以重写父类的方法并且添加自己的属性和方法。**
