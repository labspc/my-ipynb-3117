我们在 Dept2 部分学习过模块化编程，这本章我们要正式进入 OOP 学习。 OOP 这个词我在上中学的时候就已经听说，但到今天才正式进入学习。虽然，在上学期的实验中，老师也讲到 OOP 里的“封装”，但是我认为我没有完全掌握 OOP，对其中的一些概念还没有理解。

今天，参照《程序设计导论，Python语言实践》这本书，先对 OOP 进行一个深入学习，后续再参照 `MIT Software Construction 课程` ，把二者结合起来，更加深入去理解这样一种程序设计的思路。 

OOP 是如何被引出的呢？ 一个重要的概念就是：抽象。 可以说，整个程序设计，就是在运用“抽象”概念。这一点，我在学习高等数学时有很深的感触（毕竟我已经三四年没学过数学了）。如何从一个具体的东西，抽象出来一个模型，这是我们常遇到的问题。 在做数学的时候，我们常常从一个“引子”到“概念”的推出，在做程序设计的时候，我们其实干的是同样一件事。

OOP 是一种编程思想，是一种编程范式（ **编程范式（programming paradigm），所谓编程范式就是程序设计的方法论** ）。在本章中，我们主要这样一件事：使用、创建和设计高级数据类型。

## 3.2 类与对象 Class and Object

我们回顾何为数据类型？ 简单说，`值的集合+值上操作的集合=数据类型` 。

有数据类型后，我们就可以使用数据类型来创建对象。在具体讲，OOP中类和对象之前，我们要对OOP对概念做一个说明：

> 如果要用一句话来概括面向对象编程，我认为下面的说法是相当精辟和准确的。

> 面向对象编程：把一组数据和处理数据的方法组成对象，把行为相同的对象归纳为类，通过封装隐藏对象的内部细节，通过继承实现类的特化和泛化，通过多态实现基于对象类型的动态分派。

> 这句话对初学者来说可能不那么容易理解，但是我可以先为大家圈出几个关键词：对象（object）、类（class）、封装（encapsulation）、继承（inheritance）、多态（polymorphism）。

> 我们先说说类和对象这两个词。在面向对象编程中，类是一个抽象的概念，对象是一个具体的概念。我们把同一类对象的共同特征抽取出来就是一个类，比如我们经常说的人类，这是一个抽象概念，而我们每个人就是人类的这个抽象概念下的实实在在的存在，也就是一个对象。简而言之，类是对象的蓝图和模板，对象是类的实例，是可以接受消息的实体。

> 在面向对象编程的世界中，一切皆为对象，对象都有属性和行为，每个对象都是独一无二的，而且对象一定属于某个类。对象的属性是对象的静态特征，对象的行为是对象的动态特征。按照上面的说法，如果我们把拥有共同特征的对象的属性和行为都抽取出来，就可以定义出一个类。
